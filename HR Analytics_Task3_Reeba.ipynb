## Data Preparation

```python
df1=df1.drop('enrollee_id',axis=1) # dropping the column enrollee id
```

```python
df_num=df1.select_dtypes(np.number)  # creating a dataframe of numerical variables
df_num  
```

```python
df_cat=df1.select_dtypes(np.object) # creating a dataframe of categorical variables
df_cat
```

```python
df_cat_dum=pd.get_dummies(df_cat)  # creating dummy variables for categorical fields
df_cat_dum.head()
```

```python
df_merged = pd.concat ([df_num,df_cat_dum],axis=1) # merging the dataframes
df_merged
```

```python
X = df_merged.drop('target',axis=1) # dropping target column and defining X
X.head()
```

```python
y = df_merged[['target']] # defining y
y
```

```python
from sklearn.model_selection import train_test_split  # splitting the dataset into train and test datasets
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
X_train.shape
```

### 1) Logistic Regression Model (Binary Classification)

```python
from sklearn.linear_model import LogisticRegression # importing the necessary class from sklearn package of scikit learn library
```

```python
model1=LogisticRegression()    # creating an object called model1
```

```python
model1.fit(X_train,y_train)                # training the model
```

```python
score1 = model1.score(X_test,y_test)    # testing the accuracy of the model
```

### 2) K Nearest Neighbors Classification Model

```python
from sklearn.neighbors import KNeighborsClassifier #importing the necessary class from sklearn package
model2 = KNeighborsClassifier(n_neighbors = 10)
model2.fit (X_train,y_train)  # training the model
```

```python
score2 = model2.score(X_test,y_test)  # finding the accuracy of the model
```

### 3) Decision Tree Model

```python
from sklearn.tree import DecisionTreeClassifier #importing the necessary class from sklearn package
```

```python
model3= DecisionTreeClassifier()  #initiating the model
model3.fit(X_train,y_train) #training the model
```

```python
score3=model3.score(X_test,y_test)  # finding the accuracy of the model
```

### 4) Random Forest Model

```python
from sklearn.ensemble import RandomForestClassifier  # importing the necessary class from sklearn package
model4 = RandomForestClassifier() # creating an object called model4
model4.fit(X_train,y_train) # training the model 
score4=model4.score(X_test,y_test) # finding the accuracy of the model 
```

### Comparing the Models

```python
model=[['Logistic Regression',score1],['K Nearest Neighbors',score2],['Decision Tree',score3],['Random Forest ',score4]]
```

```python
pd.DataFrame(model,columns=['Model','Accuracy'])
```

```python
max_score = max(score1,score2,score3,score4)
max_score
```

```python
print ('Highest accuracy is {}' .format(max_score))  # Logistic Regression model has the highest accuracy
```
