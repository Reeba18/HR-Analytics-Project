### Q1 Are there any prominent cities where enrollees are more inclined to look for new job?

```python
plt.figure(figsize=(20,50))
a=sns.countplot(data=df2,y='city',order=df2['city'].value_counts().index)
#df2.city.value_counts().plot(kind='barh',color='orange')
plt.title("City wise Comparison of Job-Seeking Enrolees",size=20)
a.set_ylabel("City Code",size=20)



```

### Q2 Are enrolees with lower years of experience looking for a new job?

### Q4 Do enrolees looking for new jobs frequently change their jobs or spend considerable time in their last job?
